```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 200)  # x data
y = np.sin(x) + x            # y data
plt.plot(x, y) 
plt.xlabel('X Data')               # creates the plot
plt.ylabel('Y Data')               # creates the plot
plt.show()                   # this will show your plot in case your environment is not interactive
```

```python
plt.plot(x,y, color = 'y',linestyle = 'dotted', linewidth = 5,label = 'My Data')
plt.legend()
plt.show()
```
